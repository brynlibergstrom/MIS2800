#Block 1

```python
%%sql
CREATE TABLE Marketing_Brands (
  brand_id INTEGER PRIMARY KEY,
  brand_name TEXT,
  brand_type TEXT,
  brand_budget INTEGER
) STRICT;
```

#Block 3

```python
%%sql
SELECT *
FROM Marketing_Brands;
```

#Block 4

```python
%%sql
SELECT brand_name, brand_budget
FROM Marketing_Brands;
```

#Block 5

```python
%%sql
SELECT *
FROM Marketing_Brands
WHERE brand_type = 'Religious';
```

#Block 6

```python
%%sql
SELECT brand_name
FROM Marketing_Brands
WHERE brand_budget > 5000000;
```

#Block 7

```python
%%sql
SELECT *
FROM Marketing_Brands
WHERE brand_budget > 7000000;
```

#Block 8

```python
%%sql
SELECT *
FROM Marketing_Brands
WHERE brand_type = 'Snacks'
   OR brand_type = 'Beverages';
```

#Block 9

```python
%%sql
SELECT brand_name, brand_budget
FROM Marketing_Brands
ORDER BY brand_budget DESC;
```
